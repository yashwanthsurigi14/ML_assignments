```python
import statsmodels.formula.api as smf
model=smf.ols('Salary~YearsExperience',data=sl).fit()
```

```python
model.summary()
```

```python
pre=model.predict(sl.YearsExperience)
```

```python
pre
```

```python
sl
```

```python
model.resid
```

```python
model_l_rmse=np.sqrt((np.mean(np.array(sl.Salary)-np.array(pre)))**2)
model_l_rmse
```

```python
model_influence = model.get_influence()
(c, _) = model_influence.cooks_distance
```

```python
plt.stem(np.arange(len(sl)),np.round(c,3))
```

```python
(np.argmax(c),np.max(c))
```

```python
# NOW PERFROMING LOG TRANSFORMATION TO CHECK DOES IT BE A GOOD MODEL THAN SIMPLE ONE
```

```python
model_log=smf.ols("Salary~np.log(YearsExperience)",data=sl).fit()
model_log.params
```

```python
model_log.summary()
```

```python
plt.scatter(x=sl.YearsExperience,y=sl.Salary,color='red')
plt.plot(sl.YearsExperience,pre,color='black')
```

```python
model_log_pre=model_log.predict(sl.YearsExperience)
model_log_pre
```

```python
model_log.resid
```

```python
model_log_rmse=np.sqrt((np.mean(np.array(sl.Salary)-np.array(model_log_pre)))**2)
model_log_rmse
```

```python
model_influence = model_log.get_influence()
(c1, _) = model_influence.cooks_distance
```

```python
(np.argmax(c1),np.max(c1))
```

```python
plt.stem(np.arange(len(sl)),np.round(c1,3))
```

```python
# FROM BOTH THE MODELS (R_squared of SIMPLE) > (R_squared of LOG)
# THEREFORE WE CONCLUDE SIMPLE LINEAR MODEL WITH R_squared value =95.7% IS MORE EFFICENT
```
